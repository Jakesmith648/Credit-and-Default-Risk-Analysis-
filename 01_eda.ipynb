```python
df['loan_status'].value_counts()
```

Here I want to look at and explore the differnt data set values and how many each one of each category appears. 

```python
# Create binary target: 1 = defaulted, 0 = did not default
df['loan_default'] = df['loan_status'].apply(lambda x: 1 if x == 'Charged Off' else 0)

# Check distribution
df['loan_default'].value_counts(normalize=True)
```

Wanting to see what percentage of loans were payed back (O) and which were defaulted on (1).  87.9 percent of people paid back their loans and 12 percent defaulted. 

```python
df.groupby('grade')['loan_default'].mean().sort_values(ascending=False)
```

Loan Grades and there link with defaulting. 

```python
df['income_group'] = pd.cut(df['annual_inc'], bins=[0, 40000, 80000, 120000, 200000, 1_000_000])
df.groupby('income_group')['loan_default'].mean()
```

Here we can see the segmentation of loan by income.  For example I found that by segmenting the data we found that as income increases loan defaults decrease.
